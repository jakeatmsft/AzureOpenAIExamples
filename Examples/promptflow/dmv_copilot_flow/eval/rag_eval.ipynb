```python
import os
import pandas as pd
import glob

# Get a list of all JSONL files in the directory
directory = "C:/Users/jacwang/.promptflow/.runs/rag_eval_default_20240228_131905_892000/flow_artifacts"
file_paths = glob.glob(directory + "/*.jsonl")

# Read each JSONL file into a dataframe and store them in a list
dataframes = []
for file_path in file_paths:
    df = pd.read_json(file_path, lines=True)
    dataframes.append(df)

# Concatenate all dataframes into a single dataframe
combined_df = pd.concat(dataframes, ignore_index=True)

# Explode the run_info object
exploded_df = pd.json_normalize(combined_df['run_info'])

# Merge the exploded DataFrame with the original DataFrame
combined_df = pd.concat([combined_df, exploded_df], axis=1)

# Drop the original run_info column
combined_df.drop('run_info', axis=1, inplace=True)
```
